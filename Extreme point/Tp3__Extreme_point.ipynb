### Soit la s√©rie suivante :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = {'t': [1,2,3,4,5,6,7], 'CA': [120,155,125,202,180,235,240]}
df=pd.DataFrame(data)
df
```

### 1. Repr√©senter graphiquement la s√©rie par un nuage de points.

```python
plt.scatter(df['t'], df['CA'])
plt.xlabel('t')
plt.ylabel('CA')
plt.title('Nuage de points')
plt.show()
```

### 2. D√©terminer l‚Äô√©quation de la droite de tendance en utilisant la m√©thode des points extr√™mes.

```python
Y1=df.iloc[len(df)-1,1]
Y2=df.iloc[0, 1]
X2=df.iloc[0, 0]
X1=df.iloc[len(df)-1,0]
a=(Y1-Y2)/(X1-X2)
b=Y1-a*X1
print(X1,X2,Y1,Y2,a)
print(f"l‚Äô√©quation de l‚Äôajustement : Y= {a}X + {b} ")
df
```

### 3. Repr√©senter cette √©quation sur le m√™me graphique.

```python
t_line = np.linspace(1, 8, 100)
CA_line = a * t_line + b
sns.scatterplot(x=df['t'], y= df['CA'], label='Donn√©es')
sns.lineplot(x=t_line, y=CA_line, color='red', label=f'Droite de tendance: y = {a:.2f}x + {b:.2f}')
```

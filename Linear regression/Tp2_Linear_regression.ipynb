# **TP régression linéaire**

## 1. Saisie des données

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {'Heures': [1, 2, 3,4, 5], 'Note': [9, 14, 19, 24,30]}
df = pd.DataFrame(data)
df
```

## 2. Visualisation des données

```python
plt.scatter(df['Heures'], df['Note'])
plt.xlabel('Heures de révision')
plt.ylabel('Note')
plt.title('Relation entre les heures de révision et la note')
plt.show()
```

## 3. Calcul des paramètres


Linear-Regression-Formula.png

```python
X = np.array(df['Heures'])
Y = np.array(df['Note'])
n = len(X)
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X ** 2)
sum_Y2 = np.sum(Y ** 2)
a = (sum_Y*sum_X2 -sum_X*sum_XY) / (n * sum_X2 - sum_X ** 2)
b = (n*sum_XY -  sum_X*sum_Y) / (n * sum_X2 - sum_X ** 2)
print(f"l’équation de l’ajustement : Y= {a}X + {b} ")
```

## 4. Tracer la droite de l’ajustement :

```python
plt.scatter(X, Y, color='blue', label='Points de données')
plt.plot(X, a * X + b, color='red', label=f'Droite d\'ajustement : Y = {a:.2f}X + {b:.2f}')
plt.xlabel('Durée de révision (heures)')
plt.ylabel('Note d\'examen')
plt.title('Régression Linéaire : Note d\'examen , Durée de révision')
plt.legend()
plt.grid(True)
plt.show()
```

# 5.Calculer le coefficient de corrélation
Correlation-Coefficient-Formula.png

```python
#methode 1
correlation_coef1 = np.corrcoef(df['Heures'], df['Note'])[1, 0]
#methode 2
from math import *
correlation_coef2 = (n*sum_XY-sum_X*sum_Y)/sqrt((n*sum_X2-sum_X**2)*(n*sum_Y2-sum_Y**2))
print("Coefficient de corrélation:", correlation_coef1,correlation_coef2)
```
